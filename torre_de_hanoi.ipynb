```python
def torre_de_hanoi(numero_de_discos, torre_inicial, torre_auxiliar, torre_final):
  state = {
      torre_inicial: [ disco for disco in range(numero_de_discos, 0, -1)],
      torre_auxiliar: [],
      torre_final: [],
  }



  def print_state():
    for key, torre in state.items():
      text = f'{key}: ' + ' '.join(map(str, torre))
      print(text)

    print()
  def resolver_torre(n, origem, auxiliar, destino):
    if n > 0:
        resolver_torre(n - 1, origem, destino, auxiliar)

        disco = state[origem].pop()
        state[destino].append(disco)
        print_state()

        resolver_torre(n - 1, auxiliar, origem, destino)

  print_state()
  resolver_torre(numero_de_discos, torre_inicial, torre_auxiliar, torre_final)
```

```python
torre_de_hanoi(5, 'A', 'B', 'C')
```
